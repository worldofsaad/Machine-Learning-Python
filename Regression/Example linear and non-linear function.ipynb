```python
import numpy as np
import matplotlib.pyplot as pl
%matplotlib inline

print("Linear FUNCTION");
x=np.arange(-5.0, 5.0, 0.1)
y=2*(x) + 3
y_noise = 2 * np.random.normal(size = x.size)
ydata=y+y_noise

pl.plot(x,ydata,'bo')
pl.plot(x,y,'r')
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()



print("\n \n \nCUBIC FUNCTION");

x=np.arange(-5.0,5.0,0.1)
y= 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
y_data = y + y_noise
pl.plot(x,y_data,'bo')
pl.plot(x,y,'r')
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()



print("\n \n \nQUADRITIC FUNCTION 𝑌=𝑋2 ");

x=np.arange(-5.0,5.0,0.1)
y= np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
y_data = y + y_noise
pl.plot(x,y_data,'bo')
pl.plot(x,y,'r')
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()


print("\n \n \nEXPONENTIAL FUNCTION 𝑌=𝑎+𝑏𝑐𝑋 where b ≠0, c > 0 , c ≠1, and x is any real number. The base, c, is constant and the exponent, x, is a variable.");

x=np.arange(-5.0,5.0,0.1)
y= np.exp(x) #e^x
pl.plot(x,y)
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()


print("\n \n \nLogarithmic FUNCTION 𝑌=log(x) ");

x=np.arange(-5.0,5.0,0.1)
y= np.log(x)
pl.plot(x,y)
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()





print("\n \n \nSigmoidal/Logistic");

x=np.arange(-5.0,5.0,0.1)
y= 1-4/(1+np.power(3, X-2))
pl.plot(x,y)
pl.xlabel("Independent Variable")
pl.ylabel("Dependent Variable")
pl.show()
```
